```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 

tripdata_df = pd.read_csv('Resources/201908-citibike-tripdata.csv')
```

```python
# 2. Check the datatypes of your columns. 

tripdata_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.

tripdata_df['tripduration_dt']=pd.to_datetime(tripdata_df['tripduration'], unit='s', errors='ignore')
tripdata_df.sample(5)
```

```python
# 4. Check the datatypes of your columns. 

tripdata_df.dtypes
```

```python
#new column order 

new_column_order= ["tripduration_dt", "starttime", "stoptime", "start station id", "start station name", "start station latitude", "start station longitude", "end station id", "end station name", "end station latitude", "end station longitude", "bikeid", "usertype", "birth year", "gender", "tripduration"]
tripdata_df = tripdata_df[new_column_order]
tripdata_df.head()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.

tripdata_df.to_csv('Resources/201908-citibike-tripdata_dt.csv', index=False)
```
